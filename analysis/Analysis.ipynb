```python
health_data['related_drugs'].replace(to_replace=r': http[^ ]*', value='',regex=True, inplace=True)
```

```python
health_data['related_drugs']
```

```python
health_data['drug_classes'] = health_data['drug_classes'].str.split(',')
health_data = health_data.explode('drug_classes')
```

```python
health_data.head(5)
```

```python
health_data['weighted_rating'] = health_data['rating']*health_data['no_of_reviews']
health_data['normalised_weighted_score']=(health_data['weighted_rating'] - health_data['weighted_rating'].min())/(health_data['weighted_rating'].max() - health_data['weighted_rating'].min()*100)
```

```python
health_data.head()
```

```python
health_data['Exists'] = health_data['drug_name'].isin(['related_drugs']).any()
```

```python
health_data.head()
```
